```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data
X = np.load("Datasets/kryptonite-18-X.npy")
y = np.load("Datasets/kryptonite-18-y.npy")

# Calculate the correlation matrix
corr_matrix = np.corrcoef(np.concatenate((X, y.reshape(-1, 1)), axis=1), rowvar=False)

# Set the diagonal elements to 0
# np.fill_diagonal(corr_matrix, 0)

# Create the heatmap
fig, ax = plt.subplots(figsize=(22, 10))
sns.heatmap(corr_matrix, cmap='coolwarm', vmin=-0.02, vmax=0.012, center=0, annot=True, 
            xticklabels=[f"Feature {i+1}" for i in range(X.shape[1])] + ["Target"],
            yticklabels=[f"Feature {i+1}" for i in range(X.shape[1])] + ["Target"],
            ax=ax)
ax.set_title("Correlation Heatmap")
plt.show()
```
